```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_list=[5,10,15]
fig, axes=plt.subplots(1,2,sharey=True)
ax_rec=axes[1]
for n in n_list:
    data=np.loadtxt(f'time_data/recurrent_{n}.time')
    ax_rec.plot(data[:,0], (data[:,1]), label=f'n={n}')
    #ax_rec.set_yscale('log')
    ax_rec.set_xlabel(r'$|E|$')
ax_rec.legend()
ax_rec.set_title('recurrent')

# ax_reccn=axes[2]
# n_list=[10,20,30]
# for n in n_list:
#     data=np.loadtxt(f'time_data/recurrent_cn_{n}.time')
#     ax_reccn.plot(data[:,0], (data[:,1]), label=f'n={n}')
#     #ax_reccn.set_yscale('log')
#     ax_reccn.set_xlabel(r'$|E|$')
# ax_reccn.legend()
# ax_reccn.set_title('recurrent with partitioning')


ax_cut=axes[0]
data=np.loadtxt(f'time_data/cutrk.time')
ax_cut.plot(data[:,0], data[:,1])
#ax_cut.set_yscale('log')
ax_cut.set_xlabel('n')
ax_cut.set_title('cut-rank')
fig.text(0.04, 0.5, 'Run Time', va='center', rotation='vertical')
```
